### a) Implement a scatter plot version that uses the marker size aesthetic to visualize another dimension of data (up to 4D).

```python
# Load the iris dataset with the `iris_with_ints.csv` file.
data = Data('iris_with_ints.csv')

# Create a transformation object with the data.
transformation = Transformation(orig_dataset=data)

# Project the data with the headers [`sepal_length`, `petal_length`, `sepal_width`, 'petal_width].
transformation.project(['sepal_length', 'petal_length', 'sepal_width', 'petal_width'])

# Using scatter_size to recreate an plot with headers [`sepal_length`, `petal_length`, `petal_width`, 'sepal_width'].
transformation.scatter_size(x = 'sepal_length', y = 'petal_length', z = 'petal_width', size_var = 'sepal_width', title = 'Sepal Length vs Petal Length vs Petal Width')
```
