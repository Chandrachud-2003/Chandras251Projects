```python
# Making a transformation object with the iris_with_ints.csv file
data = Data('data/iris_with_ints.csv')

# Creating a transformation object with the iris_with_ints.csv file
transformation = Transformation(orig_dataset=data)

# Projecting onto the first 2 variables sepal_length, sepal_width
transformation.project(['sepal_length', 'sepal_width'])

# Plotting the original data
transformation.scatter('sepal_length', 'sepal_width', 'Original Data')

# Rotating the original data by 45 degrees
transformation.rotate_2d(45)

# Plotting the rotated data
transformation.scatter('sepal_length', 'sepal_width', 'Rotated Data')
```

# Extension Report:

This extension added a new method rotate_2d(degrees) to the Transformation class. This method takes the projected data from the last project operation and rotates it about the origin by an angle provided by the user. 

The rotate 2d method computes the rotation matrix depending on the input angle and then rotates the data using this matrix. The scatter method is then applied to the rotated data to generate a scatter plot in the original projected space. 

The code was tested by projecting the iris dataset onto the first two variables, sepal length and sepal width. The scatter method was used to plot the original and rotated data.

To put the method to the test further, other rotation angles (e.g., 10, 45, 90) can be employed to see how they affect the shape of the data. By comparing the scatter plots, one can see how the data points are positioned after rotation in relation to the original dataset.
