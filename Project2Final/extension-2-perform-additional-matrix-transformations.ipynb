# Extension Report:

### a) Normalizing by Z-score instead of Min/Max

The functions normalize)z_together() and normalize_z_separately() are used to normalize data using z-score normalization. Both algoirthms compute the data's mean and standard deviation before computing the z-scores of the data points by removing the mean and dividing by the standard deviation, and the difference between the two approaches is that normalize z together() computes the mean and standard deviation for all variables simultaneously, whereas normalize z separately() computes them separately for each variable.

### b) "Whiten" a dataset

To preprocess the data and prepare it for analysis, it is helpful to standardize it by performing a series of steps that results in a new dataset with zero mean and unit variance. This process is known as "whitening," and it can be done in a variety of ways.

For this extension, the whiten() function initially computes the mean & standard deviation across all variables. It then computes the Z-score by subtracting the mean and dividing by the standard deviation. Finally, the function centers the data around zero.

The whiten() function returns a new dataset that has been transformed to have zero mean and unit variance, which is then used to create a new Data object, which is stored as the transformation.data attribute.

To test the whiten() function, the whitened data is displayed, and an Analysis.pair plot() method is used to construct a pair plot of the normalized data. The pair plot illustrates the associations between variables in the whitened dataset. 

Overall, the whiten() function is a beneficial preprocessing step that can assist to enhance statistical model correctness and interpretability by eliminating the impact of disparities in scale and distribution between variables.


### c) Implement normalize together and separately using numpy vectorization/broadcasting.

Vectorization/broadcasting is used to implement normalize_vectorization_together() and normalize_vectorization_separately(). Both algorithms compute the global minimum and range across all variables and the local minimum and range for each variable, and then the data is scaled to a range of [0, 1] by subtracting the minimum from the data array and dividing the range by the data array.

The primary goal of these extensions is studying the impact of various normalization approaches on the dataset's projection onto a lower-dimensional space. Vectorization/broadcasting is more efficient than the preceding methods since it makes use of the CPU's parallelism capabilities.

Ultimately, these extensions offer many methods for normalizing data, and it is up to the data analyst to determine which strategy is best for their research.
