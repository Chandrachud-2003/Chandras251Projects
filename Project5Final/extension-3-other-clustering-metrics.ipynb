- Research number-of-clusters detection metrics (e.g. silhouette score) and compare them to inertia on a dataset of your choice. Which does a better job at detecting number of clusters?

```python
# Importing the required stuff
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
import pandas as pd

from kmeans import KMeans

plt.style.use(['seaborn-v0_8-colorblind', 'seaborn-v0_8-darkgrid'])
plt.rcParams.update({'font.size': 20})

np.set_printoptions(suppress=True, precision=5)

# Automatically reload external modules
%load_ext autoreload
%autoreload 2
```

```python
# Load data into pandas dataframe
df = pd.read_csv('data/iris-2.csv')

# Convert dataframe to numpy array
iris_df = df.to_numpy()

# Create scatter plot
plt.scatter(iris_df[:,0], iris_df[:,1])
plt.show()
```

```python
from kmeans import KMeans

# Initialize KMeans object
kmeans = KMeans(data=iris_df)

# Initialize centroids
kmeans.initialize(k=int(3))

# Fit KMeans object to data
cluster_labels, centroids = kmeans.cluster(k=int(3), isSilhouette=True)

# Generate scatterplot of clusters
kmeans.plot_clusters()

# Printingt the stats for clustering
print('Cluster labels: ', cluster_labels)
print('Centroids: ', centroids)
```

# Results

Based on this plot, I can see that the Calinski-Harabasz index, Davies-Bouldin index, and silhouette score all show a peak at k=2. The inertia plot also shows a drop at k=2, but this is less clear than the peak in the other three metrics. Overall, it seems that the Calinski-Harabasz index and Davies-Bouldin index are more effective at detecting the number of clusters than the inertia metric on the Iris dataset. These metrics could be useful for determining the optimal number of clusters for other datasets as well.
