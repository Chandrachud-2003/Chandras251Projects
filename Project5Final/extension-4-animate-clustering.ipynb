```python
# Importing all the required s
import os
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
import pandas as pd

from kmeans import KMeans

plt.style.use(['seaborn-v0_8-colorblind', 'seaborn-v0_8-darkgrid'])
plt.rcParams.update({'font.size': 20})

np.set_printoptions(suppress=True, precision=5)

# Automatically reload external modules
%load_ext autoreload
%autoreload 2
```

```python
# Load data into pandas dataframe
df = pd.read_csv('data/iris-2.csv')

# Convert dataframe to numpy array
iris_df = df.to_numpy()

# Create scatter plot
plt.scatter(iris_df[:,0], iris_df[:,1])
plt.show()
```

```python
from kmeans import KMeans

# Initialize KMeans object
kmeans = KMeans(data=iris_df)

# Initialize centroids
kmeans.initialize(k=int(3))

# Fit KMeans object to data
cluster_labels, centroids = kmeans.cluster(k=int(3), animate=True)

# Generate scatterplot of clusters
kmeans.plot_clusters()
```

# Results

For this extension, I created an animation that shows the K-means clustering process unfolding across iterations until the algorithm converges. The animation displays the data points as small circles, colored according to the cluster that each point belongs to, and the centroids as larger circles. The animation iteratively updates the assignment of data points to the centroids and moves the centroids to the center of their assigned points until the algorithm converges.

To create the animation, I used the matplotlib.animation library to update the plot for each iteration of the algorithm. I added an additional argument animate to the cluster function of the KMeans class, which is a boolean value that determines whether to show the animation or not. If animate=True, the function initializes a figure and axes object, sets up the animation object, and passes the figure and axes objects and the number of frames to the FuncAnimation function.

The resulting animation allows for a visual understanding of the K-means clustering process and provides insight into how the algorithm partitions the data points into clusters. By watching the animation, one can see how the algorithm works iteratively to assign points to centroids and adjust the position of the centroids until convergence.

In the context of the iris dataset used in this project, the animation provides a clear picture of how the K-means algorithm groups the different types of iris flowers based on the length and width of their petals and sepals. This information could be useful in real-world applications of the K-means algorithm to plant classification and identification.
