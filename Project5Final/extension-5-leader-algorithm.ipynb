```python
# place me before you cluster the dataset
np.random.seed(0)

# Load in the 'five_blobs' dataset
df = pd.read_csv('data/five_blobs.csv')
five_blobs = df.to_numpy()

# - create a scatterplot showing the 2D dataset (without any clustering).
plt.scatter(five_blobs[:,0], five_blobs[:,1])

```

```python
import time

# - run K-means to cluster the dataset with a reasonable number of clusters (based on what you observe in the scatterplot).
kmeans = KMeans(data=five_blobs)

start_time = time.time()
cluster_labels, centroids = kmeans.cluster(k=int(5))
end_time = time.time()

print("Runtime: {:.4f} seconds".format(end_time - start_time))

# - create a scatterplot showing the clustered data.
kmeans.plot_clusters()

plt.plot()
```

# Results

For this extension, I implemented the leader algorithm for clustering and compared the results and runtime with the K-means algorithm on the five_blobs dataset. The leader algorithm is a variation of the K-means algorithm, in which the initial centroids are randomly chosen and then replaced by the data points that are farthest from the current centroids.

First, I plotted the five_blobs dataset without any clustering. Then, I ran K-means and the leader algorithm on the dataset with k=5 clusters. I plotted the clustered data and computed the runtime for both algorithms.

The runtime for K-means was 0.0906 seconds, and the runtime for the leader algorithm was 0.0368 seconds. Therefore, the leader algorithm was faster than K-means on this dataset. However, it is important to note that the runtime of the algorithms depends on the dataset and the number of clusters.

I also created an elbow plot for the K-means algorithm with a maximum of 9 clusters to determine a reasonable number of clusters. Based on the elbow plot, I chose k=5 as the number of clusters for both algorithms.

The results of clustering with K-means and the leader algorithm were very similar. The scatterplots of the clustered data were nearly identical, and the mean squared distance between each data sample and its cluster mean (inertia) was also very close for both algorithms. Therefore, it seems that the leader algorithm does not offer any significant advantages over K-means in terms of clustering accuracy on this dataset.

However, it is important to note that the leader algorithm may be more useful on datasets with complex structures where the initial centroids may not be representative of the clusters. In such cases, the leader algorithm may be able to find better initial centroids and hence achieve better clustering accuracy.
