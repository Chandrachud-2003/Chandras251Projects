# Extension 1 - More visualizations

For this extension, I created the following additional visualizations using various types of charts and graphs:

1. Box plots: Box plots can show the distribution of each variable and identify any outliers or unusual values. They can be particularly useful for variables like price and carat weight, which may have a large range of values.

2. Histograms: Histograms can help show the distribution of a single variable. They can be used to show the distribution of carat weight, price, and other numeric variables.

3. Heatmap: A heatmap can show the correlation matrix of the dataset, allowing for an easy visual identification of strong correlations between variables.

4. 3D scatterplot: A 3D scatterplot can provide a visual representation of how three variables relate to one another. For example, we can plot the carat weight on the x-axis, the depth on the y-axis, and the price on the z-axis.

5. Density plot: A density plot can be used to show the distribution of a variable, similar to a histogram. However, it can be more useful in cases where there are multiple overlapping distributions, as it can show each distribution separately.

```python
# Importing the diamonds.csv dataset

# Importing libraries
from data import Data
import numpy as np
import pandas as pd
from analysis_extension_1 import Analysis
import matplotlib.pyplot as plt

# Fixing the diamonds csv file
diamonds_filename = 'data/diamonds.csv'

# Creating a data object
diamonds_data = Data(diamonds_filename)

# Exploring the data

# Creating an analysis object
diamonds_an = Analysis(diamonds_data)
```

```python
# Plotting boxplots for the diamonds dataset
diamonds_an.boxplot(['carat', 'price'])
plt.show
```

```python
# Plotting histograms for the diamonds dataset
diamonds_an.histogram(['carat', 'price'])
plt.show()
```

```python
# Creating 3D scatterplots for the diamonds dataset
diamonds_an.scatter3d(['carat', 'price', 'depth'])
plt.show()
```

```python
# Creating density plots for the diamonds dataset
diamonds_an.density(['carat', 'price'])
plt.show()
```
